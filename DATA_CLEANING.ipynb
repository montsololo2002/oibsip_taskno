# Missing Data Handling

```python
df_train.isnull().sum()
```

Missing values are present in four columns: 'name', 'host_name', 'last_review', and 'reviews_per_month'. The remaining columns are indicated to have no missing values, as shown by the null (0) counts.

These missing values can be handled by imputation, in which they are replaced by the mean of the dataset. However, since the columns 'name', 'host_name', and 'last_review' contain string values, they cannot be imputated as the mean function expects numerical data (float64 or int64). As such, these columns can be dropped as they are irrelevant for future analysis.

```python
df.drop(['name', 'host_name', 'last_review'], axis=1, inplace=True)
df.head()
```

Now, we can replace the missing values in the 'reviews_per_month' column with the mean.

```python
# Missing Values
data = ['reviews_per_month']
df = pd.DataFrame(df_train)

# Fill missing values with the mean
df[data] = df[data].fillna(value=df[data].mean(), inplace=False)
print("Fill with mean:\n", df)
```

```python
df_train.head()
```

For the NaN values in the 'reviews_per_month' column, we will replace them with 0, since the missing values indicate that there were no reviews.

```python
df_train.fillna({'reviews_per_month': 0}, inplace=True)
```

Now, we can verify if the missing values have been handled accordingly.

```python
df_train.isnull().sum()
```

We can verify that there are no longer missing values in the dataset.

# Outlier Detection

```python
df_train.describe()
```

```python
# Outliers
columns_to_check = ['price', 'minimum_nights', 'number_of_reviews',
                    'reviews_per_month', 'calculated_host_listings_count',
                    'availability_365']
df = pd.DataFrame(df_train)
# Calculating IQR
Q1 = df_train[columns_to_check].quantile(0.25)
Q3 = df_train[columns_to_check].quantile(0.75)
IQR = Q3 - Q1

# Defining outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtering outliers
outliers_mask = (df_train[columns_to_check] < lower_bound) | (df_train[columns_to_check] > upper_bound)

# To get the actual outlier rows:
outlier_rows = df_train[outliers_mask.any(axis=1)]

print("Outliers:\n", outlier_rows)
```

```python
ols=['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']

# Use the defined variable 'ols' instead of 'cols'
for col in ols:
    print(col)
    print('Skew :', round(df_train[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    df_train[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df_train[col])
    plt.show()
```

Since outliers are evident in all these columns, we can also apply the Z-score method to detect them.

```python
# Select numeric columns
df_numeric = df.select_dtypes(include=[np.number])

for col in df_numeric.columns:
    # Calculate z-score of the column
    z_score = np.abs(stats.zscore(df[col]))
    outliers_num = len(np.where(z_score > 3)[0])
    if outliers_num:
        print('{}: {}'.format(col, outliers_num))
```

Using Z-score to remove outliers.

```python
z_scores = np.abs(stats.zscore(df_numeric))

# DataFrame without outliers
df_wo_outliers = df[(z_scores < 3).all(axis=1)]
df_wo_outliers.shape
```
